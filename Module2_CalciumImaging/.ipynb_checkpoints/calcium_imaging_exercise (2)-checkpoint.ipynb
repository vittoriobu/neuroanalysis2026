**Why do we need source extraction if we can already see calcium traces from individual pixels?**

- Each pixel captures a mixture of signals from multiple overlapping neurons and background — it does not correspond to one single neuron.
- With 1-photon imaging, out-of-focus neurons from other focal planes further contaminate every pixel.
- Source extraction (CNMF-e) spatially demixes these overlapping contributions to isolate the activity of individual neurons.
- Averaging fluorescence over the full spatial footprint of a neuron greatly increases SNR compared to a single pixel.
- Working directly with pixels does not scale: a 400×400 ROI produces 160,000 traces, while source extraction yields a manageable number of neuron-level traces.
